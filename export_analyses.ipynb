```python
import pandas as pd
```

```python
data = pd.read_excel('exp-cust-ctry-tariff.xlsx', sheet_name = 1, header = 1 )
```

```python
data.columns
```

```python
list(data.columns)
```

```python
data.rename(columns = {'سال': 'year',
 'نام گمرک' : 'customs_name',
 'نام کشور' : 'country_name',
 'کدتعرفه': 'tarrif_code',
 'شرح تعرفه' : 'tarrif_description',
 'وزن (کیلوگرم)' : 'weight(kg)',
 'ارزش (ریال)' : 'cost(IR)',
 'ارزش (دلار)' : 'cost(dollar)'}, inplace = True)
```

```python
data
```

```python
data.groupby(by ='country_name').size().sort_values().tail(5)
```

```python
data.info()
```

```python
data.groupby(by =['customs_name', 'country_name']).max().sort_values(by = 'cost(dollar)')
```

```python
ta_de =data.groupby('tarrif_description').sum().sort_values(by = 'cost(dollar)').tail(50)
ta_de.to_excel('50_most_exportet.xlsx')
```
